```python
import pandas as pd
```

```python
df = pd.read_csv('glassdoor_jobs.csv')
```

```python
df.info()
```

```python
df.shape
```

```python
df['Salary Estimate'].unique()
```

```python
#removing the -1 salary rows
df = df[df['Salary Estimate']!='-1']
```

```python
# since we have 2 typs of salary Employer Provided Salary n Per Hour so will make seperate columns for them
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['Employer_Provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
#removing (Glassdoor est.), K$ from salary column
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0].replace('$','').replace('K',''))
```

```python
salary
```

```python
df[['Salary Estimate','hourly','Employer_Provided']]
```

```python
#removing per hour n employer provided salary from Salary Estimate
min_hr_est = salary.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
#creating min n max salary by spliting the min_hr_est on (-) n taking 1st one as min n other as max
df['min_salary'] = min_hr_est.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = min_hr_est.apply(lambda x: int(x.split('-')[1]))
df['avg_salary'] = (df.min_salary+df.max_salary)/2
```

```python
# df.head(30)
```

```python
df.info()
```

```python
# df['Company Name'].unique().tolist()
```

```python
#removing last 5 char of every company
# looks like comp name which doesn't have rating attach to its name also doesn't have rating in rating column(-1) 
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-5], axis=1)
```

```python
df['company_txt'].unique()
```

```python
df.head(20)
```

```python
# Will split on (,) n take 2nd part
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

```python
df['job_state'].value_counts()
```

```python
# cheking if an individual's job is in the Headquarters or not
# so if job_state == Headquarters return 1 else 0
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0 , axis=1)
```

```python
# df
```

```python
df['age'] = df.Founded.apply(lambda x:x if x<1 else 2021-x)
```

```python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df.python_yn.value_counts()
```

```python
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df.R_yn.value_counts()
```

```python
df['spark_yn'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df.spark_yn.value_counts()
```

```python
df['aws_yn'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df.aws_yn.value_counts()
```

```python
df['excel_yn'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel_yn.value_counts()
```

```python
#droping unamed column
df_out = df.drop(['Unnamed: 0'], axis=1)
```

```python
df_out
```

```python
#exporting it to salary_data_cleaned.csv
df_out.to_csv('salary_data_cleaned.csv', index=False)
```
